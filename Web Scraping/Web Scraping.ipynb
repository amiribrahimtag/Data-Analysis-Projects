```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
import seaborn as sns
from bs4 import BeautifulSoup
import requests
import openpyxl
```

```python
excel = openpyxl.Workbook()
print(excel.sheetnames)
sheet = excel.active
sheet.title = 'Wuzzuf Data Analyst Jobs'
print(excel.sheetnames)
sheet.append(['Job_Title','Comp_Name','Location','Job_skills'])
```

```python
pages = list(range(0,12))
for page in pages:
    url = requests.get(f'https://wuzzuf.net/search/jobs/?a=hpb&q=Data%20Analyst&start={page}')

    soup = BeautifulSoup(url.text,'html.parser')
    jobs = soup.find_all('div',class_='css-1gatmva e1v1l3u10')

    for job in jobs:
        Job_title = job.find('a',class_="css-o171kl").text
        Comp_name = job.find('a',class_="css-17s97q8").text.strip('-')
        Location = job.find('span',class_="css-5wys0k").text
        Job_skills = job.find('div',class_="css-y4udm8").text.replace("Full Time","Full Time · ").strip() #.split("·")
        full = [Job_title,Comp_name,Location,Job_skills]
        print(full)
        sheet.append(full)
        print("Scape Done")

    excel.save('June 20,2022 Jobs.xlsx')
```
